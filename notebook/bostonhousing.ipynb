## Linear Regression

```python
# call the model of linear regression
model = LinearRegression()
```

```python
# training process
model.fit(X_train,y_train)
```

```python
# predic X_train and X_test
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
# import the modul of score and accuracy
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.metrics import r2_score
```

```python
# to know mean square error
s_train = mean_squared_error(y_train,y_train_pred)
s_test = mean_squared_error(y_test,y_test_pred)
#the mean squared error,lower the value better, if it is a .0 means perfect prediction
print(f'Mean Squared error of testing set: {s_train}')
print(f'Mean Squared error of training set: {s_test}')
```

```python
# to know r2 score
sr2_train = r2_score(y_train, y_train_pred)
sr2_test = r2_score(y_test,y_test_pred)
print(f'R2 variance score of testing set: {sr2_train}')
print(f'R2 variance score of testing set: {sr2_test}')
```

## Polynomial Regression

```python
# import polynomial package
from sklearn.preprocessing import PolynomialFeatures
```

```python
# creat a polynomial regression model for the given degree=2
polyreg = PolynomialFeatures(degree=2)
```

```python
# transform the existing feature to high degree features
X_train_poly = polyreg.fit_transform(X_train)
X_test_poly = polyreg.fit_transform(X_test)
```

```python
# fit the transform features to linear regression
linreg = LinearRegression()
linreg.fit(X_train_poly, y_train)
```

```python
# predicting on training data set 
y_train_predict = linreg.predict(X_train_poly)
# predicting on testing data set
y_test_predict = linreg.predict(X_test_poly)
```

```python
# evaluating the model on train dataset
rmse_train = np.sqrt(mean_squared_error(y_train,y_train_predict))
r2_train = r2_score(y_train,y_train_predict)
print(f'R2 variance score of testing set: {rmse_train}')
print(f'R2 variance score of testing set: {r2_train}')
```
